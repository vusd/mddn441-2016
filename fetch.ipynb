```python
# all imports go at the top of the notebook
import urllib
from IPython.display import Image
from IPython.display import HTML
import os
```

```python
# make the inputs directory if it does not exist
if not os.path.exists("inputs"):
    os.mkdir("inputs")
```

```python
# download a file into the inputs directory
remote_file = "http://www.victoria.ac.nz/__data/assets/image/0004/359563/wasp-banner-white.jpg"
local_file = "inputs/wasp.jpg"
urllib.urlretrieve(remote_file, local_file)
print ("file downloaded: {}".format(local_file))
```
